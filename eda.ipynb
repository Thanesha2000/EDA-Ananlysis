```python
data=pd.read_csv('diwalidata.csv',encoding='latin-1')
data.describe()
```

```python
sns.countplot(x='Gender',data=data)#x=for the x axis and the data which we are using is sotred with variable data so ya
```

```python
ax = sns.countplot(x='Gender', data=data)

# Add labels to bars
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=True)
```

```python
salesgen=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=True)
sns.barplot(x='Gender',y='Amount',data=salesgen)
```

```python
salesage=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=True)
sns.barplot(x='Age Group',y='Amount',data=salesage)
```

```python
state=data.groupby(['State',as_index=False])['Orders'].sum().sort_values(by='Orders',ascending=True)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Orders',data=state)
```
